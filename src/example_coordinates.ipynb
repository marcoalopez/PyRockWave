# Demonstration of ``coordinates.py`` functionality

The ``coordinates.py`` file contains several functions to transform between spherical and cartesian coordinates and vice versa, as well as spherical coordinate generators.

```python
import matplotlib.pyplot as plt
import coordinates as c
```

## Test ``equispaced S2 grid``

The ``equispaced S2 grid`` function uses a modified version of the offset Fibonacci lattice algorithm to generate an approximately equispaced spherical grid.

```python
azimuths, polar = c.equispaced_S2_grid(n=100, degrees=True, hemisphere='upper')
```

```python
azimuths, polar = c.equispaced_S2_grid(n=1000)

# convert from spherical to 3d cartesian
x, y, z = c.sph2cart(azimuths, polar)

# plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, constrained_layout=True)

ax.scatter(x, y, z)
ax.axis('equal')
_ = ax.set_xlabel(r'$x$'), ax.set_ylabel(r'$y$'), ax.set_zlabel(r'$z$')
```
