# Demonstration of the functionality of the ``coordinates`` module

The ``coordinates.py`` file contains several functions to transform between spherical and cartesian coordinates and vice versa, as well as spherical coordinate generators.

```python
import matplotlib.pyplot as plt

# import required PyRockWave modules
import coordinates
```

## Test ``equispaced_S2_grid``

The ``equispaced S2 grid`` function uses a modified version of the offset Fibonacci lattice algorithm to generate an approximately equispaced spherical grid. This is the default method and works as follows:

```python
azimuths_ang, polar_ang = coordinates.equispaced_S2_grid(num_points=100, degrees=True, hemisphere='upper')
```

## Plot propagation directions (wavevectors) on a sphere surface

```python
# Use the default method: modified offset Fibonacci lattice algorithm
azimuths, polar = coordinates.equispaced_S2_grid(num_points=1000)

# convert from spherical to 3d cartesian
x, y, z = coordinates.sph2cart(azimuths, polar)

# plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, constrained_layout=True)

ax.scatter(x, y, z)
ax.axis('equal')
_ = ax.set_xlabel(r'$x$'), ax.set_ylabel(r'$y$'), ax.set_zlabel(r'$z$')
```

```python
# An alternative method: Equal-Area Partitioning method
points = coordinates.equispaced_S2_grid_eap(num_points=1000)
```

```python
fig2, ax = plt.subplots(subplot_kw={"projection": "3d"}, constrained_layout=True)

ax.scatter(points[:, 0], points[:, 1], points[:, 2])
ax.axis('equal')
_ = ax.set_xlabel(r'$x$'), ax.set_ylabel(r'$y$'), ax.set_zlabel(r'$z$')
```

```python
# compare both methods
fig3, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(7, 7), constrained_layout=True)

ax.scatter(x, y, z, label="offset Fibonacci lattice")
ax.scatter(points[:, 0], points[:, 1], points[:, 2], label="Equal-Area Partitioning")
ax.legend()

# Remove the axes panes and grid
ax.grid(False)  # Turn off the grid
ax.axis('off')  # Turn off the axis lines
# ax.view_init(elev=20, azim=30)
_ = ax.axis('equal')
```
